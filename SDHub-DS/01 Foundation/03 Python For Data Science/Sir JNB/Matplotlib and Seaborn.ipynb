```python
plt.scatter(df['Gross Sales'],df['Profit'])

plt.title("Relationship between Gross Sales and Profit")
plt.xlabel("Gross Sales")
plt.ylabel("Profit")

plt.show()
```

```python
df[['Gross Sales','Profit']].corr()
```

**Correlation is measure to check relationship between two or more numerical columns:**

- Positive: One increase, other will also increase
- Negative: One increase, other will decrease
- Almost zero: Neutral (Almost no relation or affect)

```python
df3 = df.groupby(['Product'])[['Profit']].mean()

df3
```

```python
plt.bar(df3.index,df3['Profit'])

plt.title("Average Profit by Product")
plt.ylabel("Profit")

plt.show()
```

```python
plt.barh(df3.index,df3['Profit'])

plt.title("Average Profit by Product")
plt.xlabel("Profit")

plt.show()
```

```python
# display gross sales and profit by product

df4 = df.groupby(['Product'])[['Gross Sales','Profit']].mean()

df4
```

```python
plt.bar(df4.index,df4['Gross Sales'])
plt.bar(df4.index,df4['Profit'])

plt.title("Average Profit by Product")
plt.ylabel("Profit")

plt.show()
```

```python
plt.bar(df4.index,df4['Gross Sales'],label='Gross Sales')
plt.bar(df4.index,df4['Profit'],label='Profit')

plt.title("Average Profit by Product")
plt.ylabel("Profit")

plt.legend()


plt.show()
```

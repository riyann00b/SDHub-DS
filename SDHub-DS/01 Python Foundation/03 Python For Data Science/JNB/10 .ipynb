```python
df = pd.read_csv('Dataset/Snitch_Fashion_Sales_Uncleaned.csv')

df.head()
```

```python
x = df['Profit']

plt.hist(x)

plt.title("profit Distribution")

plt.ylabel("Frequency")

plt.xlabel("Profit")

plt.show()
```

```python
df.columns
```

```python
df1 = df.groupby(['City'].dropdublicate)[['Profit']].mean()

df1
```

```python
plt.figure(figsize=(12,9))

plt.plot(df1.index, df1['Profit'])

plt.show()
```

```python
df.columns
```

```python
plt.scatter(df["Unit_Price"],df["Profit"])
plt.title("Relation Between Discount_% and Profit")
plt.show()
```

```python
df3 = df.groupby(['Product_Name'])[['Profit']].mean()

df3
```

```python

plt.figure(figsize=(30,15))

plt.barh(df3.index, df3['Profit'])

plt.title("Bar Chat")

plt.xlabel('Profit')

plt.show()
```

```python
df.columns
```

```python
df4 = df.groupby(['Product_Name'])[['Profit']].mean()

df4
```

```python
plt.pie(df4['Profit'], labels=df4.index, autopct=%%)
```

## Distributions

1. Bernoulli

```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Probability of success (1)
p = 0.6  

# Generate 1000 random trials
data = bernoulli.rvs(p, size=1000)

# Show proportion of 0s and 1s
print("Mean (Expected Value):", np.mean(data))
print("Variance:", np.var(data))

# Visualization
plt.hist(data, bins=2, edgecolor='black')
plt.xticks([0,1])
plt.title("Bernoulli Distribution")
plt.show()
```

2. Binomial

```python
from scipy.stats import binom

n = 10   # number of trials
p = 0.5  # probability of success (heads)
x = np.arange(0, n+1)

# PMF: Probability of getting exactly x successes
pmf = binom.pmf(x, n, p)

print("PMF values:", pmf)

# Plot
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

3. Normal

```python
from scipy.stats import norm

mu, sigma = 0, 1  # mean=0, std=1 (Standard Normal)
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color="red")
plt.title("Normal Distribution (mean=0, std=1)")
plt.show()
```
